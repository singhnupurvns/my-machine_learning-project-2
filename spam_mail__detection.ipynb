```python
data_cleaned["Spam"] = data_cleaned["Category"].apply(lambda x:1 if x== "spam"else 0)
data_cleaned.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data_cleaned.Message,data_cleaned.Spam,test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Correcting the pipeline definition
clf = Pipeline([
    ("vectorizer", CountVectorizer()),  # Apply CountVectorizer to the text data
    ("nb", MultinomialNB())             # Apply MultinomialNB classifier
])

# Example of fitting the model (assuming x_train and y_train are defined)
clf.fit(x_train, y_train)
```

```python
# Assuming clf is the trained pipeline, and x_test, y_test are the test data and labels
score = clf.score(x_test, y_test)

print("Accuracy score:", score)
```
